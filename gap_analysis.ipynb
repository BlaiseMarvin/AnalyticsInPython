### Gap Analysis with Continuous Variables:
***
Purpose of gap analysis: determine whether two samples of data are different. In our running example, we want to determine whether Sample 1 (salaries of female employees in the bank), is different from Sample 2 (salaries of male employees at the bank). We generally come at gap analysis in 2 steps:
- Plot the data in such a way that we can visually assess whether a gap exists. These visualizations also come in handy later when communicating the results of any formal analysis
- Conduct a formal gap analysis using statistical techniques

```python
bank = pd.read_csv('Bank.csv')
```

```python
sns.boxplot(x=bank['Salary'], y=bank['Gender'], showmeans=True);
```

from the boxplots above, you can see that males earn considerably better than females.
We can do the same kind of analysis by 'JobGrade'. But JobGrade is an integer and we did not convert it to a category variable.

```python
sns.boxplot(x=bank['Salary'],y=bank['JobGrade'].astype('category'),
            showmeans=True);
```

Gap analysis: how we can analyse the gap between 2 distributions from 2 types of populations. We can use boxplots - provide us with a visual analysis of the two distributions.

### Histograms:
Comparative histograms are a bit trickier. Given that boxplots provide much the same information, it might not be worth the effort to generate meaningful comparative histograms. However, histograms (and kernel density plots), seem to be slightly easier for some people to understand.

#### Faceted Histograms:
- We used the notion of a facet in R to create a grid of histograms. In this case, we want a grid with one column and 2 rows. The rows correspond to the different values of the Gender variable. Do this with seaborn's displot function. 

```python
sns.displot(x='Salary',col='Gender',data=bank,
            kde=True);
```

##### Overlaying Kernel Density Plots
Overlaying histograms almost never makes sense - the result is typically a mess. Here is an overlayed histogram for the bank salary data. Here is an overlayed histogram:

```python
sns.histplot(x='Salary', hue='Gender', data=bank, linewidth=0);
```

Another approach is to stack kernel density plots. Notice that some shading has been added to make the results look marginally cooler:

```python
sns.kdeplot(x='Salary', hue='Gender', data=bank, shade=True);
```

```python
sns.kdeplot(x='Salary',hue='Gender',data=bank,shade=True);
```

```python
sns.boxplot(x='Salary',y='Gender',data=bank)
```

```python
sns.displot(x='Salary',col='Gender',data=bank,kde=True)
```

There is the visual inspection of the data and there also are statistical tests. This is where we have tests such as the t-Tests. We use the t-test at this point to formally test the hypothesis that two distributions have the same sample mean (and thus are the same or at least close enough).
